```python
from bs4 import BeautifulSoup
import re
from fuzzywuzzy import fuzz
```

```python
with open("books.html") as fp:
    soup = BeautifulSoup(fp,'html.parser')
```

```python
authors=list()
books=list()
for author in soup('strong'):
    books.append(author.find_next('em').text)
    authors.append(author.text)
```

```python
def searchForBook(lines,book):
    for i in range(len(lines)):
        if fuzz.partial_ratio(book,lines[i])>60:
            print(book,i)
```

```python
for i in range(len(books)):
    searchForBook(lines,books[i])
    
```
